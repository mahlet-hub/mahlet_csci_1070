```python
import matplotlib.pyplot as plt


titanic_clean = titanic_df[['Age', 'Fare']].dropna()

bins = [0, 12, 17, 55,float('inf')]   
age_labels = ['0-12', '13-17', '18-55', '55-']
titanic_clean['age_bin'] = pd.cut(titanic_clean['Age'], bins=bins, labels=age_labels, right=False)


fig, axes = plt.subplots(2, 2, figsize=(14, 10))


selected_bins = ['0-12', '13-17', '18-55', '56-']

for i, age_bin in enumerate(selected_bins):
    row = i // 2
    col = i % 2
    fare_data = titanic_clean[titanic_clean['age_bin'] == age_bin]['Fare']
    
    axes[row, col].hist(fare_data, bins=20, color='skyblue', edgecolor='black')
    axes[row, col].set_title(f"Fare Distribution for Ages {age_bin}")
    axes[row, col].set_xlabel("Fare")
    axes[row, col].set_ylabel("Frequency")


plt.tight_layout()
plt.show()
```
